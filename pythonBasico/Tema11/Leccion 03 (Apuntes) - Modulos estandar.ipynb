# Módulos Estándar en Python

## Módulos Esenciales

### 1. copy
* Utilizado para crear copias de variables referenciadas en memoria
* Especialmente útil para colecciones y objetos

### 2. collections
* Mejora las funcionalidades de las estructuras de datos básicas
* Incluye implementaciones especializadas como colas

### 3. datetime
* Manejo de fechas y horas
* Esencial para operaciones relacionadas con el tiempo

### 4. doctest y unittest
* Módulos para crear pruebas (testing)
* Fundamentales para validar el funcionamiento del código

### 5. html, xml y json
* Manejo de estructuras de datos en diferentes formatos
* Muy utilizados en desarrollo web

### 6. pickle
* Trabajo con ficheros y objetos
* Serialización y deserialización de objetos Python

### 7. math
* Funciones matemáticas esenciales
* Operaciones matemáticas avanzadas

### 8. re (Regular Expressions)
* Expresiones regulares para búsquedas y patrones
* Especialmente útil para manipulación de cadenas

### 9. random
* Generación de contenidos aleatorios
* Útil en videojuegos y pruebas

### 10. socket
* Comunicación entre máquinas
* Creación de sistemas cliente-servidor

### 11. sqlite3
* Sistema de gestión de bases de datos relacional
* Bases de datos basadas en archivos

### 12. sys
* Información y manejo del sistema operativo
* Módulo avanzado con funcionalidades del sistema

### 13. threading
* División de procesos en subprocesos
* Programación paralela y concurrente

### 14. tkinter
* Módulo de interfaz gráfica estándar
* Creación de formularios y aplicaciones de escritorio

```python
# Ejemplo básico de importación de módulos
import datetime
import math
import random

# Fecha actual
print(f"Fecha actual: {datetime.datetime.now()}")

# Número PI
print(f"Valor de PI: {math.pi}")

# Número aleatorio entre 1 y 100
print(f"Número aleatorio: {random.randint(1,100)}")
```
