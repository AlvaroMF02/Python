## El plano cartesiano

Representa un espacio bidimensional (en 2 dimensiones), formado por dos rectas perpendiculares, una horizontal y otra vertical que se cortan en un punto. La recta horizontal se denomina eje de las abscisas o **eje X**, mientras que la vertical recibe el nombre de eje de las ordenadas o simplemente **eje Y**. En cuanto al punto donde se cortan, se conoce como el **punto de origen O**.

<img src="https://docs.hektorprofe.net/cdn/wp-content/uploads/2015/10/eje-2d-vacio.png" width="350" />

Es importante remarcar que el plano se divide en 4 cuadrantes:

<img src="https://matematicascercanas.com/wp-content/uploads/2020/02/Ejes_00-1024x819.jpg" width="350" />

## Puntos y coordenadas

El objetivo de todo esto es describir la posición de **puntos** sobre el plano en forma de **coordenadas**, que se forman asociando el valor del eje de las X (horizontal) con el valor del eje Y (vertical).

La representación de un punto es sencilla: **P(X,Y)** dónde X y la Y son la distancia horizontal (izquierda o derecha) y vertical (arriba o abajo) respectivamente, utilizando como referencia el punto de origen (0,0), justo en el centro del plano.

<img src="https://docs.hektorprofe.net/cdn/wp-content/uploads/2015/10/eje-2d.png" width="300" />


## Vectores en el plano

Finalmente, un vector en el plano hace referencia a un segmento orientado, generado a partir de dos puntos distintos. 

A efectos prácticos no deja de ser una línea formada desde un punto inicial en dirección a otro punto final, por lo que se entiende que un vector tiene longitud y dirección/sentido.


<img src="https://docs.hektorprofe.net/cdn/wp-content/uploads/2015/10/vector-plano.png" width="300" />

En esta figura, podemos observar dos puntos A y B que podríamos definir de la siguiente forma:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(5, 5)**

Y el vector se representaría como la diferencia entre las coordendas del segundo punto respecto al primero (el segundo menos el primero):
* **AB = (x2-x1, y2-y1)** => **(5-2, 5-3)** => **(3,2)** 

Lo que en definitiva no deja de ser: 3 a la derecha y 2 arriba.

````python
import math
# Completa el ejercicio aquí
class Punto:
    X=0
    Y=0
    
    # Contrusctor
    def __init__(self, x=0,y=0):
        self.X=x
        self.Y=y

    # Pasando x e y mira en q cuadrante está
    def cuadrante(self):
        if self.X == 0 and self.Y == 0:
            print("Se encuentra en 0 0")
        elif self.X > 0 and self.Y > 0:
            print("Se encuentra en el cuadrante 1")
        elif self.X < 0 and self.Y > 0:
            print("Se encuentra en el cuadrante 2")
        elif self.X < 0 and self.Y < 0:
            print("Se encuentra en el cuadrante 3")
        else:
            print("Se encuentra en el cuadrante 4")

    # Muestra los dos puntos que se han creado "graficamente"
    def vector(self,punto2):
        print(str(self) + "===>" + str(punto2))

    # Calcula la distancia que hay entre los dos puntos
    def distancia(self, x):
        return math.sqrt((x.getX() - self.X) ** 2 + (x.getY() - self.Y) ** 2)

    # Getters
    def getX(self):
        return self.X
    
    def getY(self):
        return self.Y

    # too String
    def __str__(self):
        return "("+str(self.X) +", "+ str(self.Y) +")"


class Rectangulo:
    inicial = Punto()
    final = Punto()
    base = 0
    altura = 0
    
    def __init__(self,inicial = Punto(),final = Punto()):
        self.inicial = inicial
        self.final  = final

    # Calcula la base con X2 - X
    def baseFun(self):
        self.base = self.final.getX() - self.inicial.getX()
        print("La base es: " + str(self.base))
    
    # Calcula la altura con Y2 - Y
    def alturaFun(self):
        self.altura = self.final.getY() - self.inicial.getY()
        print("La altura es: " + str(self.altura))
    
    def area(self):
        print("El area es de: " + str(self.base * self.altura))

    # too String
    def __str__(self):
        return "Inicial: "+ str(self.inicial) + " Final: " + str(self.final)
````

````python

a = Punto(2,3)
b = Punto(5,5)
c = Punto(-3,-1)
d = Punto(0,0)

print("\n")
print(a)
print(b)
print(c)
print(d)


# Consultar el cuadrante de a, c y d
print("\n ---------- CUADRANTES ----------")
a.cuadrante()
c.cuadrante()
d.cuadrante()

# Consultar vecores de AB y BA
print("\n ---------- VECTORES AB Y BA ----------")
a.vector(b)
b.vector(a)

# Consultar la distancia entra AB y BA
print("\n ---------- DISTANCIA AB Y BA ----------")
print(a.distancia(b))
print(b.distancia(a))

# Determinar cual está mas lejos de (0,0) d
print("\n ---------- PUNTO MAS LEJANO DE D (0,0) ----------")
print(a.distancia(d))
print(b.distancia(d))
print(c.distancia(d))

# Crear rectangulo con A y B
print("\n ---------- RECTANGULO CON A Y B ----------")
rect = Rectangulo(a,b)
print(rect)

# Consultar base altura y area
print("\n ---------- BASE ALTURA Y AREA ----------")
rect.baseFun()
rect.alturaFun()
rect.area()

````
