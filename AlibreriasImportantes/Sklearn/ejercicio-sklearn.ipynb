# Entrenamiento de Modelos de Machine Learning para el Dataset Iris

En este ejercicio aplicaremos diferentes modelos de machine learning al dataset Iris para comparar su rendimiento y determinar cuál es el más adecuado para este problema de clasificación.

1. Implementar diferentes modelos de clasificación usando Scikit-learn

3. Determinar el mejor modelo para este conjunto de datos

2. Preparar los datos:
