## Parte 2: Visualización de Pétalos

1. Crea un scatter plot para las medidas de los pétalos:
   - Eje X: petal length (cm)
   - Eje Y: petal width (cm)
2. Utiliza los mismos colores que antes para diferenciar especies
3. Añade los elementos decorativos necesarios (título, leyenda, etiquetas)



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')

# Crear el gráfico de dispersión
plt.scatter(
    iris[iris["species"] == "setosa"]["petal_length"],
    iris[iris["species"] == "setosa"]["petal_width"],
    c="blue", label="setosa", alpha=0.6
)
plt.scatter(
    iris[iris["species"] == "versicolor"]["petal_length"],
    iris[iris["species"] == "versicolor"]["petal_width"],
    c="green", label="versicolor", alpha=0.6
)
plt.scatter(
    iris[iris["species"] == "virginica"]["petal_length"],
    iris[iris["species"] == "virginica"]["petal_width"],
    c="red", label="virginica", alpha=0.6
)

# Etiquetas y leyenda
plt.xlabel("Longitud del pétalo (petal_length)")
plt.ylabel("Ancho del pétalo (petal_width)")
plt.title("Gráfico de dispersión: pétalo por especie")
plt.legend()

# Mostrar el gráfico
plt.show()
```
